## Pomoćne funkcije i metrike¶

```python
# Suma stepenog reda
def sum_of_serie(d,n):
    s = 0
    
    if n >= 7: #Optimizacija: Nema svrhe da za veliki stepen idemo dalje
        ns = 7  
    else:
        ns = n
        
    member = 1
    for i in range(1,ns+1): 
        member /= d 
        s += member
    
    return s

# Euklidsko rastojanje
def euclidean__distance(a,b):
    
    return math.sqrt( ((a[0] - b[0])**2) + ((a[1] - b[1])**2) )
```

```python
def metric_based_of_connection(a, b):
    
    id_a = int(a[2])
    id_b = int(b[2])

    matrix_columns = matrix_of_connections.columns
    matrix_rows = list(matrix_of_connections[' Source airport ID'])
    num_of_flights = 0
    
    if str(id_a) in matrix_columns and id_b in matrix_rows:
            num_of_flights += list(matrix_of_connections.loc[
            matrix_of_connections[' Source airport ID'] == id_b][str(id_a)])[0]
        
    if str(id_b) in matrix_columns and id_a in matrix_rows:
            num_of_flights += list(matrix_of_connections.loc[
            matrix_of_connections[' Source airport ID'] == id_a][str(id_b)])[0]
    
    
    e = euclidean__distance(a,b)
    if num_of_flights == 0:
        return e
    sum_closer = sum_of_serie(4,num_of_flights)
    
    return e * (1.0 - sum_closer)
```

```python
def light_metric(a,b): #Prvi koraci su identicni kao u prethodnom racunanju;
     #Izdvajamo karakteristike, citamo iz matrice povezanosti, racunamo euklidsko rastojanje
    id_a = int(a[2])
    id_b = int(b[2])
    matrix_columns = matrix_of_connections.columns
    matrix_rows = list(matrix_of_connections[' Source airport ID'])
    num_of_flights = 0
    if str(id_a) in matrix_columns and id_b in matrix_rows:
            num_of_flights += list(matrix_of_connections.loc[
            matrix_of_connections[' Source airport ID'] == id_b][str(id_a)])[0]
    if str(id_b) in matrix_columns and id_a in matrix_rows:
            num_of_flights += list(matrix_of_connections.loc[
            matrix_of_connections[' Source airport ID'] == id_a][str(id_b)])[0]
    e = euclidean__distance(a,b)
    
    k = 7.5
    d = 20
    
    if num_of_flights > k:
        return (e * (1 - k/d))
    else:
        return (e * (1 - num_of_flights/d))
```

## Preprocesiranje podataka

```python
features = df.columns[6:9]
df[features].head()

scaler = MinMaxScaler()
scaler.fit(df[features])  
x = pd.DataFrame(scaler.transform(df[features])) 
x.columns = features

x['Airport ID'] = df['Airport ID']
x.head()
```

```python
# Funkcija za opisivanje klastera
def describe_clusters(df, features ,n, centers): # n - broj klastera
    if 'labels' not in df.columns:
        print("Ne postoji kolona sa klasterima ili se drugacije naziva")
        return
    print(" ~ Opis dobijenih klastera: ~\n\n\n")
        
    for i in range(0,n):
        this_cluster = df.loc[ df['labels'] == i]
        num_of_members = len(this_cluster)
        if num_of_members == 0: # Ako se dogodi da imamo prazan klaster
            continue              # Nema potrebe ispisivati informacije o njemu
        print("********** Klaster {} : **********\n\n\n".format(i))
        print("Udeo: {} % ".format( int((num_of_members/whole_set) * 100) ) )
        print(" - Deskriptivne statistike: ")
        print(this_cluster.describe()[features])
        current_centroid = centers.iloc[i,]
        print("- Centroid ['Longitude', 'Latitude', 'Altitude']  :   ( {} , {} , {}) "
              .format(current_centroid['Longitude'], 
                      current_centroid['Latitude'], 
                      current_centroid['Altitude']))
        print("\n\n\n\n\n")
```

```python
# Boje za vizuelizaciju klastera
colors = ['red','deepskyblue','gold','lightgreen','palevioletred','purple','orangered',
          'blueviolet','darkblue','cadetblue','indigo','tomato','mediumaquamarine',
          'olive','c','maroon','darkseagreen','cornflowerblue','orchid','lightcoral',
          'lavender','slateblue','yellowgreen','forestgreen','antiquewhite', 'sienna',
          'darkcyan','deeppink','thistle','plum','powderblue','sandybrown','m',
          'chartreuse','lemonchiffon','mistyrose','darkmagenta','brown','peru',
          'chocolate','royalblue']

# Random generator koji ce se koristiti pri konstrukciji modela -
#     da bismo pri svakom ponovnom konstruisanju dobiji isti model

rg = Random(x = 20)
```

```python
# Omotac funkcija u kojoj konstruisemo model i primenjujemo algoritam,
# analiziramo klasterovanje i pravimo vizuelizaciju rezultata
def model_KMeans(k, metric, rg):
    est = kmeans.KMeansClusterer(num_means = k, distance = metric, rng = rg)
    
    # Nemamo fit - vec cluster. Mozemo u isto vreme i da dodelimo klastere
    # odabirom opcije 'assign_clusters'.
    df['labels'] =  est.cluster(list(x.values),assign_clusters = True) # Dodeljujemo dobijene 
    num_of_clusters = max(df['labels']) + 1
    
    centers = pd.DataFrame(est.means())
    centers = pd.DataFrame(scaler.inverse_transform(centers.iloc[:,:-1]), columns = features)
    
    plt.figure(figsize= (20,10))
    for i in range(0,num_of_clusters):
        cluster = df.loc[df['labels'] == i]
        plt.scatter(cluster['Longitude'], cluster['Latitude'], color = colors[i], label = "cluster %d"%i)
        
    plt.scatter(centers['Longitude'], centers['Latitude'], marker = 'x',
                  color = 'black', label = 'centroids')    
    plt.legend(loc = 2, prop = {'size' : 10})
    plt.show()
    
    print("Dodeljeni su klasteri. Racunanje senka koeficijenta: ")
    senka_koeficijent = silhouette_score(x, df['labels'], metric = metric)
    # Ispisujemo kvalitet klasterovanja
    print("** Dobijeni senka koeficijent je {} ".format(senka_koeficijent))
    if senka_koeficijent <= 0.0:
        print("-- Los kvalitet klasterovanja (POOR)")
    
    elif senka_koeficijent <= 0.5:
        print("-- Nije dobar kvalitet klasterovanja (FAIR)")
    
    elif senka_koeficijent < 0.7:
        print("-- Dobar kvalitet klasterovanja (GOOD)")
        
    else:
        print("-- Najbolji kvalitet klasterovanja, klasteri su gusti i dobro grupisani (GOOD)")
    
    describe_clusters(df=df,features=features,n=num_of_clusters
                          ,centers = centers)
    

    
    
    
    
    

    
```
