# Klasterovanje algoritmom **DBSCAN** nad podacima o aerodromima na osnovu njihove geografske lokacije i povezanosti letovima

   Do sad smo primenjivali algoritme klasterovanja nad podacima o aerodromima samo na osnovu geografske lokacije. Takođe smo učitali matricu povezanosti i algoritmom **DBSCAN** utvrdili komponente povezanosti i uverili smo se da su nam podaci u toj matrici retki. Značajna su nam samo polja gde su ne - nula vrednosti. Što znači da su nam iz te matrice povezanosti jedino značajni podaci ako su neka dva aerodroma povezana. Nema smisla da radimo klasično grafovsko klasterovanje i dajemo veću udaljenost aerodromima koji nisu međusobno povezani, jer kao što smo videli, većina aerodroma iz skupa nema nikakav let.
   
   Sada nam samo ostaje da vidimo kako bi tačno povezanost trebala da utiče na rastojanje između dva aerodroma. Imalo bi smisla da aerodromi koji između sebe imaju više letova budu bliži. Ako su neka dva aerodroma povezana jednim ili više letova, trebalo bi naći način da smanjujemo distancu. Pri tome za proizvoljno veliki broj letova distanca mora da ima ispravne vrednosti (tj. da ne može da se desi da je smanjimo na vrednost manju od nule).
   
   Ispod je formula koja zadovoljava prehodne uslove:
   
 $$ d (a, b) = Euclid (a , b) * [ 1 - ( \frac{1}{d} + \frac{1}{d^2} + ... + \frac{1}{d^n} ) ] $$
 
<br/>
   Gde je $ Euclid(a , b) $ euklidsko rastojanje između aerodroma $a$ i $b$, $d < 2$ konstanta koju bi trebalo odabrati (neke od vrednosti koje dolaze u obzir su $ 4, 5, 8, ...$ Ako je vrednost $d$ prevelika, povezanost gotovo da neće imati uticaja na distancu), a $n$ je broj letova između aerodroma $a$ i $b$.
   <br/><br/>
   
   ***Dodatna napomena*** : *Prethodnu formulu za uticaj povezanosti smo odabrali zbog toga što, teoretski, uticaj može da se povećava za neograničeno veliki broj letova koji može da postoji između dva aerodroma. Međutim, za velike vrednosti, već za vrednosti veće od 10 promene postaju zanemarljive, a gubitak vremena na matematičko izračunavanje je ogroman. Tako da ćemo, ipak zbog primenljivosti i optimizacije, ograničiti red polinoma. Zbog toga, pošto smo se već ograničili, definisaćemo i još jednu, jednostavniju meru na osnovu koje ćemo računati uticaj povezanosti letovima na rastojanje između aerodroma. Za iste parametre algoritama ćemo posmatrati za koju meru bolje klasteruju.* <br/><br/>
   
   Učitaćemo podatke, napraviti njihovu vizuelizaciju, uraditi preprocesiranje i potom ćemo pokušati sa vrednošću paramtera $d = 4$ . Pošto nije veliki broj aerodroma iz skupa povezan letovima, razumno je da krenemo sa istim vrednostima parametara koje smo koristili pri primeni **DBSCAN** algoritma pri klasterovanju samo na osnovu geografske lokacije, a posle eventualno možemo te vrednosti neznatno povećavati i smanjivati da vidimo kako će nam se menjati rezultati. Posle ćemo isto uraditi i sa drugom merom uticaja povezanosti na rastojanje.

## Definišemo metriku na osnovu izvedene formule

```python
# Pre toga pomocne funkcije za sumu reda ciji je clan 1/d i euklidskog rastojanja
def sum_of_serie(d,n):
    s = 0
    
    if n >= 7: #Optimizacija: Nema svrhe da za veliki stepen idemo dalje
        ns = 7  
    else:
        ns = n
        
    member = 1
    for i in range(1,ns+1): 
        member /= d 
        s += member
    
    return s


def euclidean__distance(a,b):
    
    return math.sqrt( ((a[0] - b[0])**2) + ((a[1] - b[1])**2) )
    
```

```python
def metric_based_of_connection(a, b):
    
    id_a = int(a[2])
    id_b = int(b[2])

    matrix_columns = matrix_of_connections.columns
    matrix_rows = list(matrix_of_connections[' Source airport ID'])
    num_of_flights = 0
    
    if str(id_a) in matrix_columns and id_b in matrix_rows:
       # if len(list(matrix_of_connections.loc[
      #      matrix_of_connections[' Source airport ID'] == id_b][str(id_a)])) == 1:
            num_of_flights += list(matrix_of_connections.loc[
            matrix_of_connections[' Source airport ID'] == id_b][str(id_a)])[0]
        
    if str(id_b) in matrix_columns and id_a in matrix_rows:
      #  if len(list(matrix_of_connections.loc[
       #     matrix_of_connections[' Source airport ID'] == id_a][str(id_b)])) == 1:
            num_of_flights += list(matrix_of_connections.loc[
            matrix_of_connections[' Source airport ID'] == id_a][str(id_b)])[0]
    
    
    e = euclidean__distance(a,b)
    if num_of_flights == 0:
        return e
    sum_closer = sum_of_serie(4,num_of_flights)
    
    return e * (1.0 - sum_closer)
    
```

```python
# Funkcija za opisivanje klastera
def describe_clusters(df, features ,n): # n - broj klastera
    if 'labels' not in df.columns:
        print("Ne postoji kolona sa klasterima ili se drugacije naziva")
        return
    print(" ~ Opis dobijenih klastera: ~\n\n\n")
    
    noise_cluster =  df.loc[ df['labels'] == -1]
    noise_num_of_members = len(noise_cluster)
    if noise_num_of_members != 0:
        print("********** Klaster suma (eng. Noise) : **********\n\n\n")
        print("Udeo: {} % ".format( int((noise_num_of_members/whole_set) * 100) ) )
        print(" - Deskriptivne statistike: ")
        print(noise_cluster.describe()[features])
        print("\n\n\n\n\n")
        
    for i in range(0,n):
        this_cluster = df.loc[ df['labels'] == i]
        num_of_members = len(this_cluster)
        if num_of_members == 0: # Ako se dogodi da imamo prazan klaster
            continue              # Nema potrebe ispisivati informacije o njemu
        print("********** Klaster {} : **********\n\n\n".format(i))
        print("Udeo: {} % ".format( int((num_of_members/whole_set) * 100) ) )
        print(" - Deskriptivne statistike: ")
        print(this_cluster.describe()[features])
        print("\n\n\n\n\n")
```

## Primenjujemo algoritam **DBSCAN** sa našom definisanom metrikom

```python
colors = ['red','deepskyblue','gold','lightgreen','palevioletred','purple','orangered',
          'blueviolet','darkblue','cadetblue','indigo','tomato','mediumaquamarine',
          'olive','c','maroon','darkseagreen','cornflowerblue','orchid','lightcoral',
          'lavender','slateblue','yellowgreen','forestgreen','antiquewhite', 'sienna',
          'darkcyan','deeppink','thistle','plum','powderblue','sandybrown','m',
          'chartreuse','lemonchiffon','mistyrose','darkmagenta','brown','peru',
          'chocolate','royalblue']
def model_DBSCAN(df,x,eps,min_samples, metric):
    est = DBSCAN(eps = eps,min_samples = min_samples, metric = metric) # Pravimo instancu modela
    
    est.fit(x) # Treniramo model
    
    df['labels'] = est.labels_ # Dodeljujemo dobijene klastere
    
    print("Dodeljeni su klasteri. Racunanje senka koeficijenta: ")
    senka_koeficijent = silhouette_score(x, est.labels_, metric = metric)
    # Ispisujemo kvalitet klasterovanja
    print("** Dobijeni senka koeficijent je {} ".format(senka_koeficijent))
    if senka_koeficijent <= 0.0:
        print("-- Los kvalitet klasterovanja (POOR)")
    
    elif senka_koeficijent <= 0.5:
        print("-- Nije dobar kvalitet klasterovanja (FAIR)")
    
    elif senka_koeficijent < 0.7:
        print("-- Dobar kvalitet klasterovanja (GOOD)")
        
    else:
        print("-- Najbolji kvalitet klasterovanja, klasteri su gusti i dobro grupisani (GOOD)")
    num_of_clusters = max(df['labels']) + 1
    describe_clusters(df= df, n= num_of_clusters,features= features)
    
    # Iscrtavamo klastere na mapi
    
    figure = plt.figure(figsize= (20,9.5))
    for i in range(-1,num_of_clusters):
        
        if i == -1: # Da li trenutno crtamo šum
            label = "noise"
            color = "lightgray"
        else:
            label = "cluster %d"%i
            color = colors[i]
        
        # Elementi koji pripadaju klasteru koji trenutno crtamo
        current_cluster = df[df['labels'] == i]
        plt.scatter(current_cluster['Longitude'], current_cluster['Latitude'], 
                   color = color, label = label)
    plt.legend()
    plt.show()
```

```python
# Prva procena: eps = 0.05 , minimalan broj suseda = 100
model_DBSCAN(df = df,x=x,eps = 0.05, min_samples= 100, metric = metric)
```

```python
# Nova procena: eps = 0.05, minimalan broj suseda = 90

model_DBSCAN(df = df, x = x, eps = 0.05, min_samples = 88, metric = metric)
```

```python
# Skoro sve tačke su ostale šum.
# Smanjujemo dalje minimalan_broj_suseda za isto epsilon
# Nova procena: eps = 0.1, minimalan_broj_suseda = 100

model_DBSCAN(df = df, x = x, eps = 0.05, min_samples = 80, metric = metric)
```

```python
# Možemo da primetimo nezanemarljivo povećanje senka-koeficijenta.
# Na osnovu te mere nam je klasterovanje nesto bolje, ali opet nije blizu dobrog.

# Minimalan broj suseda je bio dovoljno mali da neki granični elementi postanu elementi
# jezgra i time se dva klastera (aerodroma iz Afrike i Evrope) spoje u jedan, ali opet ne dovoljno mali da bi neki element
# sa severa Azije postao element jezgra.
# Svejedno, još ćemo smanjivati minimalan_broj_suseda da vidimo šta će dalje da se menja.

model_DBSCAN(df = df, x = x, eps = 0.05, min_samples = 68, metric = metric)
```

## Jednostavnija mera uticaja na rastojanje

$$ {lowerDistance} = \frac{k}{d} * e $$

  Gde je $e$ euklidsko rastojanje između dva aerodroma, $k$ broj letova između aerodroma (ili gornja granica, ako broj je broj letova premašuje), $d$ je opšta konstanta koju bi trebalo pravilno odabrati. Ako je previše velika, poremetiće se poredak između distanci, a ako je suviše mala, povezanost neće imati preveliki uticaj.
